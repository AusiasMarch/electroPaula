```python
import folium
import geopandas as gpd
from shapely.geometry import Point, Polygon
```

```python
location = [41.581389, 1.620833]
bounds_points = [ (1.581405, 41.60138), (1.657555, 41.600512), (1.661109, 41.561733), (1.581422, 41.559295)]
bounds_poly = Polygon(bounds_points)
bounds_gpd = gpd.GeoDataFrame(gpd.GeoSeries(bounds_poly), columns=['geometry'])
```

```python
sf = gpd.read_file('shape_files/MUC_CLAS.shp')
sf
```

```python
type(sf)
```

```python
sf_igualada = gpd.overlay(sf, bounds_gpd, how='intersection')
sf_igualada
```

```python
m = folium.Map(
    location=location,
    tiles='Stamen Toner',
    zoom_start=12
)

folium.GeoJson(
    sf_igualada,
    name='shape_file'
).add_to(m)

m
```
