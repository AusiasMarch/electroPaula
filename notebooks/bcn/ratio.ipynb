```python
import os
import folium
import geopandas as gpd
```

```python
location=[41.395, +2.15]
```

```python
barris_geo = os.path.join('bcn-geodata', 'barris', 'barris_geo.json')
barris_topo = os.path.join('bcn-geodata', 'barris', 'barris_topo.json')
```

```python
barris_gpd = gpd.read_file(barris_geo)
barris_gpd['Ratio'] = barris_gpd['Homes']/barris_gpd['Dones']
```

```python
barris_gpd
```

```python
m = folium.Map(
    location=location,
    tiles='Stamen Toner',
    zoom_start=12
)

folium.GeoJson(
    barris_geo,
    name='geojson'
).add_to(m)

folium.TopoJson(
    open(barris_topo),
    'objects.barris_geo',
    name='topojson'
).add_to(m)

folium.LayerControl().add_to(m)

m
```

```python
m = folium.Map(
    location=location,
    tiles='OpenStreetMap',
    zoom_start=12
)

m.choropleth(
    geo_data=barris_gpd,
    name='choropleth',
    data=barris_gpd,
    columns=['C_Barri', 'Ratio'],
    key_on='feature.id',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Ratio Homes/Dones'
)

for i,row in barris_gpd.iterrows():
    folium.Marker([row['geometry'].centroid.y, row['geometry'].centroid.x], 
                  popup='Homes: {}<br>Dones: {}<br>Ratio : {}'.format(row['Homes'], 
                                                                       row['Dones'], 
                                                                       row['Ratio']),
                  icon=folium.Icon(color='red', icon='user')).add_to(m)

m
```
